```python
# Zaimportowanie potrzebnych bibliotek wraz z plikiem zawierającym dane do zadania
import pandas as pd
import numpy as np
df = pd.read_csv('data.csv', sep=';')
df
```

### Dodanie kolumn

```python
df.insert(1, "Height", [170, 188, 159, 174, 182, 177], True)
df.insert(3, "Hair color", ["black", "blond", "blond", "grey", "brown", "black"])
df
```

### Usunięcie kolumny

```python
df = df.drop("Hair color", axis=1)
```

 ###Dodanie wiersza

```python
df = df.append({"Name" : "Bill", "Height" : 201, "Age" : 53}, ignore_index=True)
df
```

### Dodanie kilku wierszy naraz

```python
listOfSeries = [pd.Series(["Hank", 184, 43, 115], index=df.columns),
                pd.Series(["Colin", 192, 32, 104], index=df.columns)]
df = df.append(listOfSeries, ignore_index=True)
df
```

### Usunięcie wiersza

```python
# Usunięcie wiersza z id 8 oraz 0
df = df.drop([0, 8])
df
```

```python
# Obliczenie na każdej kolumnie liczbowej mediany, średniej, maximum, minimum
median = df.median(skipna=True, numeric_only=True, axis=0)
print("Median:\n",median, "\n")

mean = df.mean(skipna=True, numeric_only=True, axis=0)
print("Mean:\n",mean, "\n")

median = df.median(skipna=True, numeric_only=True, axis=0)
print("Median:\n",median, "\n")

maximum = df.max(skipna=True, numeric_only=True, axis=0)
print("Maximum:\n",maximum, "\n")

minimum = df.min(skipna=True, numeric_only=True, axis=0)
print("Median:\n",minimum, "\n")

standard_deviation = df.std(skipna=True, numeric_only=True, axis=0)
print("Standard deviation:\n",standard_deviation, "\n")
```

```python
df.fillna(df.mean(), inplace=True)
df
```
